```python
import pandas as pd
import numpy as np
v1=np.array([1,3,6,np.nan,7,1,np.nan,9,15])
v2=np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
v3=np.array([np.nan,12,5,6,14,7,np.nan,2,31])
df=pd.DataFrame({"V1":v1,
                "V2":v2,
                "V3":v3})
df
```

```python
df.dropna()
#bir tane bile eksik deger olan gözlemler silindi
```

```python
df.dropna(how="all",axis=0)
#tüm degerleri eksik olan gözlem birimini sildik
```

```python
df["V1"].fillna(value=0,inplace=True)
```

```python
df.dropna(axis=1)
#değişken degerlerinden en az 1 tanesi bile eksik ise sil
```

```python
df["sil_beni"]=np.nan
#yeni colums ekledik tüm degerleri eksik deger olan
```
