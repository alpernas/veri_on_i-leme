```python
import pandas as pd
import numpy as np
v1=np.array([1,3,6,np.nan,7,1,np.nan,9,15])
v2=np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
v3=np.array([np.nan,12,5,6,14,7,np.nan,2,31])
df=pd.DataFrame({"V1":v1,
                "V2":v2,
                "V3":v3})
df
```

```python
#sayısal degişkenlerde atama
```

```python
df["V1"].fillna(df["V1"].mean())
```

```python
df.apply(lambda x:x.fillna(x.mean()),axis=0)#birinci yol
#değişkenlerin eksik degerlerine ortalama atması yaptık
```

```python
#2.yol
df.fillna(df.mean()[:])
```
