```python
#hızlı çözüm
import pandas as pd
import numpy as np
v1=np.array([1,3,6,np.nan,7,1,np.nan,9,15])
v2=np.array([7,np.nan,5,8,12,np.nan,np.nan,2,3])
v3=np.array([np.nan,12,5,6,14,7,np.nan,2,31])
df=pd.DataFrame({"V1":v1,
                "V2":v2,
                "V3":v3})
df
```

```python
df.isnull()#eksik degerleri true - false olarak gösterir
```

```python
df.isnull().sum()#degişkenlerdeki toplam eksik degerleri gösterir
```

```python
df.notnull().sum()#eksik deger olamayanalrın toplamını gösterir
```

```python
df.isnull().sum().sum()#toplam eksik deger sayısı
```

```python
#en az bir tane eksik deger olanları seçme işlemi yapar
df[df.isnull().any(axis=1)]
```

```python
df[df.notnull().all(axis=1)]#hepsi dolu olanları göster
```

```python
df[df["V1"].notnull() & df["V2"].notnull() & df["V3"].notnull()]
```
